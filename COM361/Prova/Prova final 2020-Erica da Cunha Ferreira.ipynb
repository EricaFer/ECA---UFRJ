## Problema 1: 
Considere o problema de publicidade em 3 mídias (canais) diferentes, para dez grupos demográficos. A matriz $P$ especificada abaixo possui 10 linhas, uma para cada grupo demográfico e três colunas, uma para cada mídia. $R_{ij}$ é o número de visualizações (em unidades de mil) do grupo $i$, por real gasto em publicidade no canal $j$. Seja $g$ o vetor em $R^3$ de gastos nos canais, então $v = Pg$ representa o vetor de número de visualizações em cada grupo demográfico. Suponha que o valor desejado $v^{des}$ é $10^3{\mathbf 1}$ (ou seja, deseja-se atingir um milhão de visualizações em cada grupo), $\mathbf 1$ é o vetor de 1s em $R^{10}$. Calcule o vetor de gastos $g^*$ que melhor atinge este alvo $v^{des}$. Calcule o vetor de gastos na presença de um orçamento limitado em R\$1400 (mil e quatrocentos reais).

$$
P = \pmatrix{0.97 & 1.86 & 0.41\\
1.23 & 2.18 & 0.53\\
0.80 & 1.24 & 0.62\\
1.29 & 0.98 & 0.51 \\
1.10 & 1.23 & 0.69\\
0.67 & 0.34 & 0.54 \\
0.87 & 0.26 & 0.62 \\
1.10 & 0.16 & 0.48\\
1.92 & 0.22 & 0.71 \\
1.29 & 0.12 & 0.62}
$$


### Solução do problem 1:

#### Rascunho 

<br><br>



```julia
# Resolução do Problema em Julia

# Lembrando que linha(i) são os grupos
# e as colunas(j) são os canais

P = [0.97  1.86  0.41
     1.23  2.18  0.53
     0.80  1.24  0.62
     1.29  0.98  0.51 
     1.10  1.23  0.69
     0.67  0.34  0.54 
     0.87  0.26  0.62 
     1.10  0.16  0.48
     1.92  0.22  0.71 
     1.29  0.12  0.62]

m = Model(with_optimizer(Ipopt.Optimizer, print_level = 0))

# Gasto tem que ser maior que 0 | 3 canais
@variable(m, g[1:3] >= 0)

# Equivalente a um milhão - em P: 1 = 1000 | 10 grupos
@variable(m, v[1:10] >= 1000)

#@variable(m, v[1:10] == 1000)

# Para cada grupo ter exatamente 1000 views,
# por incrível que pareça, fica mais caro, custa 14002
# Por isso coloquei maior ou igual, porque qual agencia 
# de publicidade não quer diminuir o custo e aumentar view?
# Enfim, achei a opção mais viável

# Soma dos gastos de todos os canais não pode ultrapassar orçamento
@constraint(m, sum(g) <= 1400)

# Aqui é o pulo do gato:
# O número de views por grupo é a soma de:
# Número de views por real por grupo e por canal
# Multplicado por gasto no canal
@constraint(m, [i in 1:10], sum(P[i,j] .* g[j] for j in 1:3) .== v[i]) 

# Objetivo é minimizar o custo
@objective(m, Min, sum(g))

optimize!(m)

# Arrendondando porque dinheiro n tem tantas casas decimais
println("Custo =  ", round(objective_value(m),digits = 2))
println()
println("---------------Investimento-------------")
println()
println("Canal 1: ", round(JuMP.value.(g[1]),digits = 2))
println("Canal 2: ", round(JuMP.value.(g[2]),digits = 2))
println("Canal 3: ", round(JuMP.value.(g[3]),digits = 2))
```

## Problema 2: 

A e B são posições e velocidades iniciais de dois barcos num lago. Os barcos possuem a mesma dinâmica dada por:
\begin{align}
x(t+1) &= x(t) + (1/3600)v(t) \\
v(t+1) & = u(t) + v(t)
\end{align}
sendo $x(t),v(t), u(t)$ posição, velocidade e aceleração no instante $t$. Os barcos
- deveriam se encontrar em $t= 60$ (o local de encontro não é importante, apenas o instante)
- Encontre uma sequência de acelerações que atinge o encontro, minimizando a energia total utilizado por ambos, representado pela função objetivo $T = \sum_{i=1}^{60} \|u^A (t)\|^2 + \sum_{i=1}^{60} \|u^B (t)\|^2$
- Aplique uma regularização $L_{\infty}$ a esta sequência, com parâmetro $\lambda = 0.5$ e compare os resultados.

```julia
# Momento de Encontro
t = 60

m = Model(with_optimizer(Ipopt.Optimizer, print_level = 0))

# Variáveis do barco A
@variable(m, A_x[1:2, 1:t]) # Posição
@variable(m, A_v[1:2, 1:t]) # Velocidade
@variable(m, A_u[1:2, 1:t]) # Aceleração

# Variáveis do barco B
@variable(m, B_x[1:2, 1:t]) # Posição
@variable(m, B_v[1:2, 1:t]) # Velocidade
@variable(m, B_u[1:2, 1:t]) # Aceleração

# Condições iniciais - Pontos determinados a partir do cálculo anexado
# OBS: levando em conta que x_a - x_b = 0.5
@constraint(m, A_x[:, 1] .== [0.5, 0])
@constraint(m, B_x[:, 1] .== [0 , 0])
@constraint(m, A_v[:, 1] .== [0 ,20])  # Velocidade de A é 20 km/h Norte(y+)
@constraint(m, B_v[:, 1] .== [30, 0])  # Velocidade de A é 30 km/h Leste(x+)

# Codificando o movimento dos barcos

# x(t + 1) = x(t) + v(t)
@constraint(m,[i in 2:t], A_x[:, i] .== A_x[:, i-1] + A_v[:, i-1]/3600)
@constraint(m,[i in 2:t], B_x[:, i] .== B_x[:, i-1] + B_v[:, i-1]/3600)
# v(t+1) = v(t) + u(t)
@constraint(m,[i in 2:t], A_v[:, i] .== A_v[:, i-1] + A_u[:, i-1])
@constraint(m,[i in 2:t], B_v[:, i] .== B_v[:, i-1] + B_u[:, i-1])

# Tem que estar na mesma posição no t = 60
@constraint(m, A_x[:, t] .== B_x[:, t])

# Minimizando a energia
@objective(m, Min, sum(A_u.^2) + sum(B_u.^2))

optimize!(m)

println("Energia Mínima =  ", objective_value(m))

#Pegando os vatores de posição para depois comparar
x_B = JuMP.value.(B_x);
x_A = JuMP.value.(A_x);
```

```julia
# Plotando gráfico

x_A_ = x_A[1, :]
y_A = x_A[2, :]

x_B_ = x_B[1, :]
y_B = x_B[2, :]

figure(figsize = (12,5))

plot(x_A_, y_A,label = "Barco A")
plot(x_B_, y_B,label = "Barco B", "r")

legend(loc = "best")

ylabel("y")
xlabel("x")

grid(true)

title("Caminho Barcos");
```

```julia
# Momento de Encontro
t = 60

m = Model(with_optimizer(Ipopt.Optimizer, print_level = 0))

# Variáveis do barco A
@variable(m, A_x[1:2, 1:t]) # Posição
@variable(m, A_v[1:2, 1:t]) # Velocidade
@variable(m, A_u[1:2, 1:t]) # Aceleração

# Variáveis do barco B
@variable(m, B_x[1:2, 1:t]) # Posição
@variable(m, B_v[1:2, 1:t]) # Velocidade
@variable(m, B_u[1:2, 1:t]) # Aceleração

# Condições iniciais - Pontos determinados a partir do cálculo anexado
# OBS: levando em conta que x_a - x_b = 0.5
@constraint(m, A_x[:, 1] .== [0.5, 0])
@constraint(m, B_x[:, 1] .== [0 , 0])
@constraint(m, A_v[:, 1] .== [0 ,20])  # Velocidade de A é 20 km/h Norte(y+)
@constraint(m, B_v[:, 1] .== [30, 0])  # Velocidade de A é 30 km/h Leste(x+)

# Codificando o movimento dos barcos

# x(t + 1) = x(t) + v(t)
@constraint(m,[i in 2:t], A_x[:, i] .== A_x[:, i-1] + A_v[:, i-1]/3600)
@constraint(m,[i in 2:t], B_x[:, i] .== B_x[:, i-1] + B_v[:, i-1]/3600)
# v(t+1) = v(t) + u(t)
@constraint(m,[i in 2:t], A_v[:, i] .== A_v[:, i-1] + A_u[:, i-1])
@constraint(m,[i in 2:t], B_v[:, i] .== B_v[:, i-1] + B_u[:, i-1])

# Tem que estar na mesma posição no t = 60
@constraint(m, A_x[:, t] .== B_x[:, t])

# ∞-normalização (para n dar negativo)
@variable(m, r)
@constraint(m, A_u .<= r  )
@constraint(m, -r .<= A_u )
@constraint(m, B_u .<= r  )
@constraint(m, -r .<= B_u )

# Criando expressões para função objetivo ficar mais limpa
@expression(m, sum_A_u, sum(A_u.^2))
@expression(m, sum_B_u, sum(B_u.^2))

# Minimizando a energia - Agora com regularização
@objective(m, Min, sum_A_u + sum_B_u + 0.5*r)

optimize!(m)

println("Energia Mínima =  ", objective_value(m))

#Pegando os vatores de posição para depois comparar
x_B_r = JuMP.value.(B_x);
x_A_r = JuMP.value.(A_x);
```

```julia
# Comparando os 2 casos

x_A_r_ = x_A_r[1, :]
y_A_r = x_A_r[2, :]

x_B_r_ = x_B_r[1, :]
y_B_r = x_B_r[2, :]

figure(figsize = (12,5))

# Barcos questão a
plot(x_A_, y_A,label = "Barco A")
plot(x_B_, y_B,label = "Barco B", "r")

# Barcos questão b
plot(x_A_r_, y_A_r,label = "Barco A - Reg")
plot(x_B_r_, y_B_r,label = "Barco B - Reg")

legend(loc = "best")

ylabel("y")
xlabel("x")

grid(true)

title("Caminho Barcos");
```
